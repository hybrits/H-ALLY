```python
import requests
from bs4 import BeautifulSoup
import csv
import time
from urllib.parse import urljoin

def get_headers():
    return {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
        'Accept-Language': 'en-US,en;q=0.5',
        'Accept-Encoding': 'gzip, deflate, br',
        'DNT': '1',
        'Connection': 'keep-alive',
        'Upgrade-Insecure-Requests': '1',
    }

def scrape_forum(base_url):
    print(f"Accediendo a la URL base: {base_url}")
    headers = get_headers()
    try:
        response = requests.get(base_url, headers=headers)
        response.raise_for_status()
    except requests.RequestException as e:
        print(f"Error al acceder a la URL base: {e}")
        print(f"Contenido de la respuesta: {response.text[:500]}...")
        return []

    soup = BeautifulSoup(response.text, 'html.parser')
    thread_links = soup.find_all('a', class_='thread-title--gtm')

    print(f"Encontrados {len(thread_links)} enlaces de hilos")

    data = []

    for i, link in enumerate(thread_links, 1):
        thread_url = urljoin(base_url, link['href'])
        print(f"Procesando hilo {i}/{len(thread_links)}: {thread_url}")

        try:
            thread_response = requests.get(thread_url, headers=headers)
            thread_response.raise_for_status()
        except requests.RequestException as e:
            print(f"Error al acceder al hilo: {e}")
            continue

        thread_soup = BeautifulSoup(thread_response.text, 'html.parser')

        # Buscar el título en diferentes elementos
        title_elements = thread_soup.find_all(['h1', 'h2', 'h3'], class_=['p-title-value', 'thread-title', 'MessageCard__thread-title'])
        if title_elements:
            title = title_elements[0].text.strip()
            print(f"Título encontrado: {title}")
        else:
            print("No se encontró el título del hilo")
            print("Elementos h1 encontrados:")
            for h1 in thread_soup.find_all('h1'):
                print(f"- {h1.text.strip()} (clases: {h1.get('class', [])})")
            title = "Título no encontrado"

        content_element = thread_soup.find('div', class_='message-content js-messageContent')
        if content_element:
            content = content_element.find('div', class_='bbWrapper')
            if content:
                post_content = content.text.strip()
                print(f"Contenido encontrado (longitud: {len(post_content)} caracteres)")
                print(f"Primeros 100 caracteres del contenido: {post_content[:100]}...")
            else:
                print("No se encontró el contenido del post")
                post_content = "Contenido no encontrado"
        else:
            print("No se encontró el elemento del contenido del mensaje")
            post_content = "Contenido no encontrado"

        data.append({'title': title, 'content': post_content})
        time.sleep(2)  # Pausa de 2 segundos entre solicitudes para ser más respetuoso con el servidor

    return data

def save_to_csv(data, filename):
    if not data:
        print("No hay datos para guardar en CSV")
        return

    print(f"Guardando {len(data)} entradas en {filename}")
    with open(filename, 'w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=['title', 'content'])
        writer.writeheader()
        for row in data:
            writer.writerow(row)
    print("Datos guardados exitosamente")

# Uso del script
base_url = "https://www.techsupportforum.com/forums/overclocking.273/"
scraped_data = scrape_forum(base_url)
save_to_csv(scraped_data, 'foro_data.csv')
```
